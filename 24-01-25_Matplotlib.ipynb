```python
import matplotlib.pyplot as plt

# Data
lang = ['Python', 'Java', 'Ruby', 'C++']
sizes = [40.0, 30.0, 10.0, 20.0]  # Numerical values for the pie chart

# Plotting the pie chart
plt.pie(sizes, labels=lang, colors=['skyblue', 'lavender', 'yellow', 'olive'], autopct='%1.1f%%',startangle=90)

#autopct='%1.1f%%':Displays with 1 decimal place
#startangle =90 :rotates the start of the pie chart to 90 degree
# Add a title
plt.title("Programming Language Usage")
plt.legend()

# Display the plot
plt.show()
```

```python
#Identify outliers in marks (box plot)
import pandas as pd
import matplotlib.pyplot as plt

# Load the School_Performance dataset
school_data = pd.read_csv('School_Performance.csv')  # Update the file path if necessary

# Reshape the data to have 'Subject' and 'Marks' columns
md = school_data.melt(id_vars=['Student_ID'], var_name='Subject', value_name='Marks')

# Create a boxplot using matplotlib

subjects = md['Subject'].unique()
data = [md[md['Subject'] == s]['Marks'] for s in subjects]

plt.boxplot(data, patch_artist=True,  labels=subjects)
plt.title('Identify outliers in marks (box plot)')
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('GlobalLandTemperaturesByCity.csv')

#Group by country and calculate the mean average temperature
country_avg = data.groupby('Country')['AverageTemperature'].mean().reset_index()

#Remove the rows with missing values in temperature 
country_avg = country_avg.dropna()
#sort the countries by average temperature in descending order
top_coutries = country_avg.sort_values(by='AverageTemperature', ascending=False).head(10)

plt.figure(figsize=(15, 6))
plt.bar(top_coutries['Country'],top_coutries['AverageTemperature'],color='lavender',edgecolor='black')

plt.title("Top 10 Countries by Average Temperature", fontsize=14)
plt.xlabel("Countries", fontsize=18)
plt.ylabel("Average Temperature (°C)", fontsize=16)
plt.xticks(rotation=45)
plt.grid(True,axis='y')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Histogram for Total Bill Amount
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="skyblue")
plt.title("Distribution of Total bills")
plt.show()
```

```python
#Kernel Density Estimation (KDE) visualizes the probability density of a variable 
sns.kdeplot(data=tips,x="total_bill",fill=True,color="purple")
plt.title("KDE Plot for Total Bills")
plt.show()
```

```python
#Average total bill by day
sns.barplot(data=tips,x="day",y="total_bill",palette="viridis")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box Plot of Total Bills by Day and Gender ")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter Plot of Total Bill vs.Tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="green")
plt.title("Line Plot of Party Size vs. Tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

```python
#Heatmaps display data as a color-coded matrix
#Correlation heatmap
numeric_tips=tips.select_dtypes(include=["number"])
correlation_matrix=numeric_tips.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('train.csv')

subset=df[["Age", "Fare", "Pclass","Survived"]]

# Select relevant columns for the pairplot
sns.pairplot(subset,  hue="Survived", palette="deep",diag_kind='kde')

# Title
plt.suptitle("Pairwise Relationships between Age, Fare, and Passenger", y=1.02)

# Show the plot
plt.show()
```

```python
numeric_train=df.select_dtypes(include=["number"])
correlation_matrix=numeric_train.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Correlation MatrixHeatmap")
plt.show()
```
